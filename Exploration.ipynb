```
import pandas as pd
import numpy as np
```

```
df=pd.read_csv("https://raw.githubusercontent.com/gaurisharma360/MEDICINES/main/Data.xlsx%20-%20Data%201.csv")
```

```
df.head(20)
```

```
#(df.columns).isnull().sum()
```

```
#dealing with null colums
null_columns=df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```
print(df[df.isnull().any(axis=1)][null_columns].head(10))
```

```
#I use DataFrame.any with parameter axis=1 to check at least one True in row by DataFrame.isna with boolean indexing:

df1 = df[df.isna().any(axis=1)]
```

```
df1.head(30)
```

```
df1["medicineType"].value_counts()
```

```
df1['pack_form'] = pd.np.where(df1.medicineType.str.contains("TABLET"), "Strip",
                   pd.np.where(df1.medicineType.str.contains("SYRUP"), "Bottle",
                   pd.np.where(df1.medicineType.str.contains("INJECTION"), "Injection", 
                   pd.np.where(df1.medicineType.str.contains("CAPSULE"),"Bottle",
                   pd.np.where(df1.medicineType.str.contains("GEL"),"Tube",
                   pd.np.where(df1.medicineType.str.contains("LIQUID"),"Bottle","--" ))))))
```
